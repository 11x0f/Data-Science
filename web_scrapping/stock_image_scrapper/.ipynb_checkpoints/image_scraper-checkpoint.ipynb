```python
import requests
from selenium import webdriver
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm 
```

```python
browser = webdriver.Edge()
```

```python
link = "https://stock-pictures.netlify.app"
```

```python
browser.get(link)
```

```python
soup = BeautifulSoup(browser.page_source, "html.parser")
```

```python
data = []
for sp in tqdm(soup.find_all("div", class_ = "container")):
    #print(sp)
    img_link = sp.find("img").get("src")
    
    tags = sp.find("div", class_="tags").text[7:].split(",")
    
    interactions = sp.find("div", class_ ="likes-comments").find_all("span")
    likes = interactions[0].text.split(" ")[0]
    comments = interactions[1].text.split(" ")[0]
    data.append([img_link, tags, likes, comments])
```

```python
data = pd.DataFrame(data, columns=["image link", "tags", "likes", "comments"])
```

```python
data.to_csv("data.csv", index=False)
```
