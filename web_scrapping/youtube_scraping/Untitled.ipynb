# Scrapping from each page

```python
import numpy as np
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
```

```python
df = pd.read_csv("data.csv")
```

```python
browser = webdriver.Edge()
```

```python
data = []
for link in tqdm(df["Video Link"]) :
    browser.get(link)

    time.sleep(5)
    
    soup = BeautifulSoup(browser.page_source, "html.parser")
    try:
        likes = soup.find("button", class_="yt-spec-button-shape-next yt-spec-button-shape-next--tonal yt-spec-button-shape-next--mono yt-spec-button-shape-next--size-m yt-spec-button-shape-next--icon-leading yt-spec-button-shape-next--segmented-start").find(
        "div", class_="yt-spec-button-shape-next__button-text-content").text
    except: likes=np.nan

    try:
        description = ""
        for sp in soup.find_all("span", class_="yt-core-attributed-string--link-inherit-color"):
            desc = sp.text
            description = description + desc
    except: description=np.nan

    data.append([likes, description])
```

```python
df_ = pd.DataFrame(data, columns=["likes", "description"] )
```

## scraping remaining videos 

```python
new_data = []
for i in tqdm(range(1522, df["Video Link"].count()) ):
    browser.get(df["Video Link"][i])

    time.sleep(5)
    
    soup = BeautifulSoup(browser.page_source, "html.parser")
    try:
        likes = soup.find("button", class_="yt-spec-button-shape-next yt-spec-button-shape-next--tonal yt-spec-button-shape-next--mono yt-spec-button-shape-next--size-m yt-spec-button-shape-next--icon-leading yt-spec-button-shape-next--segmented-start").find(
        "div", class_="yt-spec-button-shape-next__button-text-content").text
    except: likes=np.nan

    try:
        description = ""
        for sp in soup.find_all("span", class_="yt-core-attributed-string--link-inherit-color"):
            desc = sp.text
            description = description + desc
    except: description=np.nan
    new_data.append([likes, description])
```

```python
df__ = pd.DataFrame(new_data, columns=["likes", "description"] )
```

## merging both 

```python
df_merged = pd.concat([df_, df__], ignore_index=True)
len(df_merged)
```
