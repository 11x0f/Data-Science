```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
link = "https://www.youtube.com/@GeeksforGeeksVideos/videos"
prefix = "https://www.youtube.com/"
```

```python
browser = webdriver.Edge()
```

```python
browser.get(link)
```

### html page data

```python
soup = BeautifulSoup(browser.page_source, "html.parser")
```

```python
sp = soup.find("ytd-two-column-browse-results-renderer")
```

```python
sp1 = sp.find("ytd-rich-grid-renderer")
```

```python
sp2 = sp1.find("div", id = "contents")
```

```python
sp3 = sp2.find_all("ytd-rich-item-renderer")
```

## extrating data

```python
data = []
for x in sp3:
    #title, video link, image link, view count, upload date
    try:
        title = x.find("a", id="video-title-link").get("title")
    except: title = np.nan

    try:
        video_link = prefix + x.find("a").get("href")
    except: video_link = np.nan
        
    try:
        img_link = x.find("yt-image").find("img").get("src").split("?")[0]
    except: img_link= np.nan

    try:
        metadata = x.find_all("span", class_ = "inline-metadata-item")
        view_count = metadata[0].text
        upload_date= metadata[1].text
    except: 
        view_count= np.nan
        upload_date = np.nan

    try:
        duration = x.find("badge-shape").text
    except: duration =np.nan
    
    data.append([title, video_link, img_link, view_count, upload_date, duration])
```

```python
data
```

```python
len(data)
```

```python
df = pd.DataFrame(data, columns=["Title", "Video Link", "Tumbnail Link", "View Count", "Upload History", "Duration"] )
```
