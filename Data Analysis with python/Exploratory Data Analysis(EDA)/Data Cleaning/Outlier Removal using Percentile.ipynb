```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("weight-height.csv")
df
```

```python
sns.distplot(df["Height"])
```

```python
sns.boxplot(df["Height"])
```

```python
minLimit = df["Height"].quantile(0.01)
maxLimit = df["Height"].quantile(0.99)
print(minLimit," ",maxLimit)
```

```python
df[(df["Height"]<minLimit)|(df["Height"]>maxLimit)]
```

# Trimming

```python
df_new = df[(df["Height"]>minLimit) & (df["Height"]<maxLimit)]
```

```python
sns.boxplot(df_new)
```

# Capping

```python
df["Height"] = np.where(df["Height"]>maxLimit, maxLimit, np.where(df["Height"]<minLimit, minLimit, df["Height"]))
```
