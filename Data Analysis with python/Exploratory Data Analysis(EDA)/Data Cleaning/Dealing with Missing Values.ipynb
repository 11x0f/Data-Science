```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('../AB_NYC_2019.csv')
df.isnull().sum()
```

# Deleting columns with missing data

```python
df1 = df.copy()
df1.head()
df1.isnull().sum()
```

```python
df1.drop("last_review", axis=1, inplace=True)
df1.isnull().sum()
```

# # Delete the rows with missing values

```python
df2= df.copy()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace = True)
df2.isna().sum()
```

## Imputation - numerical data

```python
df3=df.copy()
```

```python
df3.head()
```

```python
meanValue = df3["reviews_per_month"].mean()
df3["reviews_per_month"].fillna(meanValue, inplace=True)
df3.head()
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy="mean")
impute.fit(df3.iloc[:,13:14].values)
df3.iloc[:,13:14]= impute.transform(df3.iloc[:,13:14].values)
df3.iloc[:,13:14]
```

```python
df3.iloc[:,13:14]
```

# Imputation - Categorical Data

```python
df4=df.copy()
```

```python
df4.isnull().sum()
```

```python
myval = df["last_review"].value_counts().index[0]
df4["last_review"].fillna(myval, inplace=True)
df4.isnull().sum()
```

```python
impute1 = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
impute1.fit(df4.iloc[:,1:2].values)
df4.iloc[:,1:2]=impute1.transform(df4.iloc[:,1:2].values)
df4.isna().sum()
```

```python
df5=df.copy()
df5.isna().sum()
```

```python
df5['last_review']
```

### Introducing new category

```python
df5["last_review"].fillna("Not Received", inplace=True)
df5['last_review']
```
