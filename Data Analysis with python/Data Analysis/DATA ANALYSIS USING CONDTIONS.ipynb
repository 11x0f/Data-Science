```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("googleplaystore.csv")
df.head(2)
```

```python
df.isna().sum()
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy ="mean")
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)
df=df.dropna()
df.isna().sum()
```

```python
df.head(2)
```

## HOW MANY FREE APPS ARE THERE IN ART_AND_DESIGN?

```python
df_processed = df[df["Category"]=="ART_AND_DESIGN"]
df_processed[df_processed["Type"]=="Free"]
print(len(df_processed))
```

## HOW MANY APPS ARE THERE IN ART_AND_DESIGN WITH RATING MORE THAN 4.5?

```python
df_processed = df[df["Category"]=="ART_AND_DESIGN"]
df_processed = df_processed[df_processed["Rating"]>4.5]
print(len(df_processed))
```

## HOW MANY FREE APPS IN FAMILY WITH RATING MORE THAN 4.5 ?

```python
df_processed = df[df["Category"]=="FAMILY"]
df_pr = df_processed[df_processed["Rating"]>4.5]
df_pr = df_pr[df_pr["Type"]=="Free"]
print(len(df_pr))
```

## LIST ALL FREE APPS IN FAMILY WITH RATING MORE THAN 4.5?

```python
df_processed = df[df["Category"]=="FAMILY"]
df_pr = df_processed[df_processed["Rating"]>4.5]
df_pr = df_pr[df_pr["Type"]=="Free"]
df_pr["App"]
result = df_pr["App"].values
```
