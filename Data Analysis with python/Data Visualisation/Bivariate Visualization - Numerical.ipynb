```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

# SCATTER PLOT

```python
sns.scatterplot(data=df.head(), x="age", y="fare", hue="survived")
```

```python
data = sns.load_dataset("tips")
data.head()
```

```python
sns.scatterplot(data=data, x="total_bill", y="tip", hue="sex")
```

```python
sns.scatterplot(data=data, x="total_bill", y="tip", hue="sex", style="smoker")
```

# JOINT PLOT 

```python
sns.jointplot(data=data, x="total_bill", y="tip")
```

```python
sns.jointplot(data=data, x="total_bill", y="tip", hue="sex")
```

```python
sns.jointplot(data=data, x="total_bill", y="tip", kind="reg")
```
