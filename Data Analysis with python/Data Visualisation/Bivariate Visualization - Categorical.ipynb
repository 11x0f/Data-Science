```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head(2)
```

```python
sns.countplot(data=df, x="survived", hue="class")
```

# HEAT MAP

```python
x= pd.crosstab(df["survived"], df["class"])
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, annot=True, fmt="0.1f", cmap="coolwarm")
```

```python
sns.barplot(data=df, x="class", y="age", hue="sex")
```

# STRIP PLOT

```python
sns.stripplot(data=df, x="class", y="age", hue="sex")
```
